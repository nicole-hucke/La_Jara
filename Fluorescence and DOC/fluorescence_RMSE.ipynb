```python
import numpy as np

# load data from csv file
data = np.genfromtxt('SM22_fluoro.csv', delimiter=',')

# assign varaibles
fi = data[1:,1]
fdom = data[1:,7]

# calculate the RMSE
rmse = np.sqrt(np.mean((fi - fdom)**2))

print("The RMSE is:", rmse)
```

```python
import matplotlib.pyplot as plt

# create scatter plot of fi vs fdom
plt.scatter(fi, fdom, s=10, c='blue', alpha=0.5)

# add 1:1 line
x = np.linspace(np.min(fi), np.max(fi), 100)
y = np.linspace(np.min(fdom), np.max(fdom), 100)
plt.plot(x, y, c='red')

# add RMSE label
plt.text(np.max(fi)- 0.05, np.min(fdom) + 0.2, 'RMSE = {:.2f}'.format(rmse), fontsize=12)

# set axis labels and title
plt.xlabel('Fluorescence Index')
plt.ylabel('fDOM')
plt.title('Fluorescence Index vs fDOM')

# show the plot
plt.savefig('fdom_fi_RMSE.png', dpi=300)
plt.show()
```

```python
# calculate correlation coefficient and R-squared value
corr_coeff = np.corrcoef(fi, fdom)[0, 1]
r_squared = corr_coeff**2

# calculate and plot fitted line
fit = np.polyfit(fi, fdom, 1)
m, b = fit
fit_fn = np.poly1d(fit)

# plot data and fitted line
plt.scatter(fi, fdom, s=10, c='blue', alpha=0.5)
plt.plot(fi, fit_fn(fi), c='black')

# add labels and title
plt.xlabel('Fluorescence Index')
plt.ylabel('fDOM')
plt.title('Fluorescence Index vs. fDOM\ny = {:.2f}x + {:.2f}\nCorrelation Coefficient: {:.2f}\nR-squared: {:.2f}'.format(m, b, corr_coeff, r_squared))

# display plot
plt.savefig('r2andcorrelation_fluoro.png',dpi=300,bbox_inches='tight')
plt.show()
```
