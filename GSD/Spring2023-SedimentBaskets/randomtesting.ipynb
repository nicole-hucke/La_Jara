```python
# Create two example DataFrames
data1 = {'A': [1, 2, 3], 'B': [4, 5, 6]}
data2 = {'C': [2]}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Choose the column from df1 and the value from df2
column_to_subtract = 'A'  # Column in df1
value_to_subtract = df2.at[0, 'C']  # Value from df2

# Subtract the value from the chosen column in df1
df1[column_to_subtract] = df1[column_to_subtract] - value_to_subtract

# Display the updated DataFrame
print(df1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame (replace with your own data)
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [2, 3, 4, 3, 2],
    'upper_limit': [3, 4, 5, 4, 3],
    'lower_limit': [1, 2, 3, 2, 1]
})

# Extract data from the DataFrame
x = data['x']
y = data['y']
upper_limit = data['upper_limit']
lower_limit = data['lower_limit']

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the area between upper and lower limits as a filled region
plt.fill_between(x, lower_limit, upper_limit, color='lightgray', alpha=0.7)

# Plot the data line in a dark color
plt.plot(x, y, color='blue', label='Data')

# Customize labels and titles
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Data with Standard Deviation Limits')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrames (replace with your actual DataFrames)
df = pd.DataFrame({'T25A': [1, 2, 3, 4, 5]}, index=[10, 25, 50, 75, 90])
upper_limit = pd.DataFrame({'T25A': [2, 3, 4, 5, 6]}, index=[10, 25, 50, 75, 90])
lower_limit = pd.DataFrame({'T25A': [0, 1, 2, 3, 4]}, index=[10, 25, 50, 75, 90])

# Create the plot
fig, ax1 = plt.subplots()

# Plot the line plot with percentiles on the Y-axis and 'T25A' on the X-axis
ax1.plot(df['T25A'], df.index, label='A (closed)')

# Fill between lower and upper limits
for idx in df.index:
    x = df.loc[idx, 'T25A']
    y_lower = lower_limit.loc[idx, 'T25A']
    y_upper = upper_limit.loc[idx, 'T25A']
    ax1.fill_betweenx([idx, idx], x, y_upper, alpha=0.6, color='blue')
    ax1.fill_betweenx([idx, idx], x, y_lower, alpha=0.6, color='blue')

plt.xlabel("X-axis Label")
plt.ylabel("Percentiles (Y-axis)")
plt.title("Fill Between with Percentiles on Y-axis")

plt.show()
```
