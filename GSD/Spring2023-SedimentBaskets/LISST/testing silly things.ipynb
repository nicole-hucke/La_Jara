```python
from scipy.stats import ks_2samp

# Example arrays (replace these with your actual data)
array1 = [2, 2, 3, 4, 5, 8, 9, 10, 11, 12]
array2 = [100, -10000, 0, 89, 1]

# Perform the Kolmogorov-Smirnov test
statistic, p_value = ks_2samp(array1, array2)

# Print the results
print(f"KS Statistic: {statistic}")
print(f"P-value: {p_value}")

# Interpret the results
if p_value < 0.05:
    print("The two samples are likely from different distributions.")
else:
    print("There is not enough evidence to suggest that the two samples are from different distributions.")
```

```python
import pandas as pd
from scipy.stats import kruskal
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Example data (replace with your actual data)
group1 = [23, 21, 18, 25, 27]
group2 = [29, 32, 27, 34, 28]
group3 = [17, 19, 22, 16, 20]

# Combine data into a DataFrame
df = pd.DataFrame({
    'Group': ['Group1'] * len(group1) + ['Group2'] * len(group2) + ['Group3'] * len(group3),
    'Value': group1 + group2 + group3
})

# Perform Kruskal-Wallis test
kruskal_stat, kruskal_p_value = kruskal(group1, group2, group3)
print(f"Kruskal-Wallis Statistic: {kruskal_stat}")
print(f"P-value: {kruskal_p_value}")

# Perform Tukey's HSD test as a post-hoc test
tukey_result = pairwise_tukeyhsd(df['Value'], df['Group'])
print("\nTukey's HSD Test Results:")
print(tukey_result)
```
