```python
#Import libraries 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

#Import Data
#Groundwater Wells (Water Table)
gw2 = pd.read_csv('GW2_SPRING2022.csv').iloc[1150:6300]
gw3 = pd.read_csv('GW3_SPRING2022.csv').iloc[1150:6300]
gw4 = pd.read_csv('GW4_SPRING2022.csv').iloc[1150:6300]
gw6 = pd.read_csv('GW6_SPRING2022.csv').iloc[1150:6300]
gw7 = pd.read_csv('GW7_SPRING2022.csv').iloc[1150:6300]
gw8 = pd.read_csv('GW8_SPRING2022.csv').iloc[1150:6300]
gw9 = pd.read_csv('GW9_SPRING2022.csv').iloc[1150:6300]
#Temperature Probes (Hyporheic Flow)
p4 = pd.read_csv('Probe4_SP22_144hr.csv').iloc[863:4865]
p6 = pd.read_csv('Probe6_SP22_144hr.csv').iloc[863:4865]
p25 = pd.read_csv('Probe25_SP22_144hr.csv').iloc[863:4865]
#Discharge Data
piezo = pd.read_csv('P2-C_Spring2022.csv')
depth = pd.read_csv('Spring2022_SmoothedDepth.csv')

#Change tile of date in all columns
p4.rename(columns={'Time': 'date'}, inplace=True)
p6.rename(columns={'Time': 'date'}, inplace=True)
p25.rename(columns={'Time': 'date'}, inplace=True)
gw3.rename(columns={'Date/time': 'date'}, inplace=True)
gw4.rename(columns={'Date/time': 'date'}, inplace=True)
gw6.rename(columns={'Date/time': 'date'}, inplace=True)
gw7.rename(columns={'Date/time': 'date'}, inplace=True)
gw8.rename(columns={'Date/time': 'date'}, inplace=True)
gw9.rename(columns={'Date/time': 'date'}, inplace=True)
```

```python
#Making a date-time format so in can be easy to plot
from datetime import datetime as dt

#Piezometer
date = []
for fecha in depth['Time']:
    date_time_obj = dt.strptime(fecha, '%m/%d/%Y %H:%M')
    date.append(date_time_obj)
new_dates = pd.DataFrame(date)
depth['new_dates'] = date

date = []
for fecha in piezo['Time']:
    date_time_obj = dt.strptime(fecha, '%m/%d/%Y %H:%M')
    date.append(date_time_obj)
new_dates = pd.DataFrame(date)
piezo['new_dates'] = date

#P4
date = []
for fecha in p4['date']:
    date_time_obj = dt.strptime(fecha, '%Y-%m-%d %H:%M:%S')
    date.append(date_time_obj)
new_dates = pd.DataFrame(date)
p4['new_dates'] = date

#P6
date = []
for fecha in p6['date']:
    date_time_obj = dt.strptime(fecha, '%Y-%m-%d %H:%M:%S')
    date.append(date_time_obj)
new_dates = pd.DataFrame(date)
p6['new_dates'] = date

#P25
date = []
for fecha in p25['date']:
    date_time_obj = dt.strptime(fecha, '%Y-%m-%d %H:%M:%S')
    date.append(date_time_obj)
new_dates = pd.DataFrame(date)
p25['new_dates'] = date

#GW3
date = []
for fecha in gw3['date']:
    date_time_obj = dt.strptime(fecha, '%m/%d/%Y %H:%M')
    date.append(date_time_obj)
new_dates = pd.DataFrame(date)
gw3['new_dates'] = date

#GW4
date = []
for fecha in gw4['date']:
    date_time_obj = dt.strptime(fecha, '%m/%d/%Y %H:%M')
    date.append(date_time_obj)
new_dates = pd.DataFrame(date)
gw4['new_dates'] = date

#GW6
date = []
for fecha in gw6['date']:
    date_time_obj = dt.strptime(fecha, '%m/%d/%Y %H:%M')
    date.append(date_time_obj)
new_dates = pd.DataFrame(date)
gw6['new_dates'] = date

#GW7
date = []
for fecha in gw7['date']:
    date_time_obj = dt.strptime(fecha, '%m/%d/%Y %H:%M')
    date.append(date_time_obj)
new_dates = pd.DataFrame(date)
gw7['new_dates'] = date

#GW8
date = []
for fecha in gw8['date']:
    date_time_obj = dt.strptime(fecha, '%m/%d/%Y %H:%M')
    date.append(date_time_obj)
new_dates = pd.DataFrame(date)
gw8['new_dates'] = date

#GW9
date = []
for fecha in gw9['date']:
    date_time_obj = dt.strptime(fecha, '%m/%d/%Y %H:%M')
    date.append(date_time_obj)
new_dates = pd.DataFrame(date)
gw9['new_dates'] = date
```

```python
# PLOTTING P25 with GW3 and GW4
fig, ax1 = plt.subplots()

ax1.set_xlabel('Date')
ax1.set_ylabel('Groundwater Level (cmH2O)', color='tab:blue')
ax1.plot(gw3['new_dates'], gw3['CompPressure[cmH2O]'], color='tab:blue', linestyle="-", label='GW3')
ax1.plot(gw4['new_dates'], gw4['CompPressure[cmH2O]'], color='tab:cyan', linestyle="-", label='GW4')
ax1.tick_params(axis='y')
plt.xticks(rotation=0)
#ax1.set_position([box.x0, box.y0 + box.height * 0.1, box.width, box.height * 0.9])
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.25),fancybox=True, ncol=3)

from matplotlib.dates import DateFormatter

date_form = DateFormatter("%m-%d")
ax1.xaxis.set_major_formatter(date_form)
ax2 = ax1.twinx()  # initiate a second axes that shares the same x-axis

ax2.set_ylabel('Hyporheic flux ($m^3$/s)')  # we already handled the x-label with ax1
ax2.plot(p25['new_dates'], p25['Pressure Temp 1'], linestyle='dashed', color='orange', label='20 cm')
ax2.plot(p25['new_dates'], p25['T3'], linestyle="dashed", color='red', label='50 cm')
ax2.plot(p25['new_dates'], p25['Pressure Temp 2'], linestyle="dashed", color='green', label='80 cm')
ax2.tick_params(axis='y')
ax2.xaxis.set_major_formatter(date_form)
ax2.set_ylim([-0.0002, 0.0002])
#ax1.set_position([box.x0, box.y0 + box.height * 0.1, box.width, box.height * 0.9])
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15),fancybox=True, ncol=3)

plt.title("Groundwater at 3&4 and hyporheic flux at P25 (What used to be P3)")
plt.grid()
sns.set_style("whitegrid")

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
