## End Members Signatures

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
import seaborn as sns
from pathlib import Path

data = pd.read_csv('endmembers_data.csv')
samples = pd.read_csv('watersample_data.csv')
samples_raw = pd.read_csv('watersample_data.csv')
samples_all = pd.read_csv('watersample_data_all.csv')
```

Delineating areas of each source signature

```python
# delineating areas (using convex hulls)
def plot_convex_hull(points, ax, color, alpha=0.2):
    if len(points) > 2:  # convexhull requires at least 3 points
        hull = ConvexHull(points)
        for simplex in hull.simplices:
            ax.plot(points[simplex, 0], points[simplex, 1], color + '--')
        # fill the convex hull
        ax.fill(points[hull.vertices,0], points[hull.vertices,1], color=color, alpha=alpha)
```

### Plotting end members

C/N vs d13C

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 9))

# scatter plot for each source with different markers and colors
markers = ['o', 's', 'D', '^', 'v', '*']
colors = ['b', 'g', 'r', 'c', 'm', 'y']
for i, source in enumerate(data['Source'].unique()):
    subset = data[data['Source'] == source]
    plt.scatter(subset['C/N'], subset['d13C'], label=source, s=100, marker=markers[i], color=colors[i])

# scatter plot for the water samples
water_markers = ['8', 'X', 'H', '<', '>', '^']
for i, source in enumerate(samples['Source'].unique()):
    subset = samples[samples['Source'] == source]
    plt.scatter(subset['C/N'], subset['d13C'], label=source, s=100, marker=water_markers[i], color='k')

# Label the axes
plt.xlabel('C/N Ratio')
plt.ylabel('13C ()')
plt.title('End Members - CN vs 13C')
plt.legend(loc='best', frameon=True)

# using the delineating function 
for i, source in enumerate(data['Source'].unique()):
    subset = data[data['Source'] == source]
    points = np.array(subset[['C/N', 'd13C']])
    plot_convex_hull(points, plt, colors[i], alpha=0.2)

# ensure the 'plots' directory exists
plots_dir = Path('plots')
plots_dir.mkdir(parents=True, exist_ok=True)
# save the figure
plt.savefig(plots_dir / 'endmembers_CN_vs_d13C.png')
plt.show()
```

```python
sns.set(style="whitegrid")
fig, axs = plt.subplots(1, 2, figsize=(16, 8))  # Create two subplots vertically

# Define markers and colors
markers = ['o', 's', 'D', '^', 'v', '*']
colors = ['b', 'g', 'r', 'c', 'm', 'y']
water_markers = ['8', '^', 'v', '<', 'X', '*', '>', 'D', 'd', 'p']

# Function to plot end members
def plot_end_members(ax):
    for i, source in enumerate(data['Source'].unique()):
        subset = data[data['Source'] == source]
        ax.scatter(subset['C/N'], subset['d13C'], label=source, s=100, marker=markers[i], color=colors[i])
        points = np.array(subset[['C/N', 'd13C']])
        plot_convex_hull(points, ax, colors[i], alpha=0.2)

# Plot end members on both subplots
plot_end_members(axs[0])
plot_end_members(axs[1])

# Plot storm samples on the first subplot
for i, source in enumerate(samples_all['Source'].unique()):
    if source.startswith('storm'):
        subset = samples_all[samples_all['Source'] == source]
        axs[0].scatter(subset['C/N'], subset['d13C'], label=source, s=100, marker=water_markers[i % len(water_markers)], color='k')

# Plot other samples on the second subplot
for i, source in enumerate(samples_all['Source'].unique()):
    if not source.startswith('storm'):
        subset = samples_all[samples_all['Source'] == source]
        axs[1].scatter(subset['C/N'], subset['d13C'], label=source, s=100, marker=water_markers[i % len(water_markers)], color='k')

# Set axis labels and titles
axs[0].set_title('Storm Samples - C/N vs 13C')
axs[1].set_title('Non-Storm Samples - C/N vs 13C')
for ax in axs:
    ax.set_xlabel('C/N Ratio')
    ax.set_ylabel('13C ()')
    ax.legend(loc='best', frameon=True)

    # Set axis bounds based on 'data'
    x_min, x_max = data['C/N'].min(), data['C/N'].max()
    y_min, y_max = data['d13C'].min(), data['d13C'].max()
    padding = 0.3
    x_range = x_max - x_min
    y_range = y_max - y_min
    ax.set_xlim(x_min - padding * x_range, x_max + padding * x_range)
    ax.set_ylim(y_min - padding * y_range, y_max + padding * y_range)

plt.tight_layout()
plots_dir = Path('plots_all_water_samples')
plots_dir.mkdir(parents=True, exist_ok=True)
# save the figure
plt.savefig(plots_dir / 'endmembers_CN_vs_d13C.png', dpi=300)
plt.show()
```
