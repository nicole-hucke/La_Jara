## Carbon end members for individual storms

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial import ConvexHull
import seaborn as sns
from pathlib import Path
import matplotlib.cm as cm

# importing data
data = pd.read_csv('endmembers_data.csv')
storm2 = pd.read_csv('storm_csvs/st2.csv')
storm3 = pd.read_csv('storm_csvs/st3.csv')
storm4_down = pd.read_csv('storm_csvs/st4_down.csv')
storm4_up = pd.read_csv('storm_csvs/st4_up.csv')
storm5_down = pd.read_csv('storm_csvs/st5_down.csv')
storm5_up = pd.read_csv('storm_csvs/st5_up.csv')
storm6 = pd.read_csv('storm_csvs/st6.csv')
storm7_down = pd.read_csv('storm_csvs/st7_down.csv')
storm7_up = pd.read_csv('storm_csvs/st7_up.csv')
```

Function for delineating areas of each source signature

```python
def plot_convex_hull(points, ax, color, alpha=0.2):
    hull = ConvexHull(points)
    for simplex in hull.simplices:
        ax.plot(points[simplex, 0], points[simplex, 1], '--', color=color)
    # Fill the convex hull
    ax.fill(points[hull.vertices, 0], points[hull.vertices, 1], color=color, alpha=alpha)
```

## Plotting end members

Storm 2 - Aug 3th 2022

```python
sns.set(style="whitegrid")
fig, ax = plt.subplots(figsize=(10, 9))

# Scatter plot for each source with different markers and colors
markers = ['o', 's', 'D', '^', 'v', '*']
colors = ['b', 'g', 'r', 'c', 'm', 'y']
for i, source in enumerate(data['Source'].unique()):
    subset = data[data['Source'] == source]
    ax.scatter(subset['%C'], subset['d13C'], label=source, s=100, marker=markers[i], color=colors[i])

# Using the delineating function  
for i, source in enumerate(data['Source'].unique()):
    subset = data[data['Source'] == source]
    points = np.array(subset[['%C', 'd13C']])
    plot_convex_hull(points, ax, colors[i], alpha=0.2)

# Scatter plot for the water samples with gradient colors based on 'order'
cmap = cm.get_cmap('copper')  # Use a grayscale colormap
orders = storm2['order']
norm = plt.Normalize(orders.min(), orders.max())
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])

for i, source in enumerate(storm2['Source'].unique()):
    subset = storm2[storm2['Source'] == source]
    colors = cmap(norm(subset['order']))
    ax.plot(subset['%C'], subset['d13C'], color='sienna', linestyle=':', linewidth=2) # connecting lines
    ax.scatter(subset['%C'], subset['d13C'], label=source, s=100, marker='^', color=colors) # actual samples


# Label the axes
ax.set_xlabel('%C')
ax.set_ylabel('δ13C (‰)')
ax.set_title('Storm 2 - %C vs δ13C')
ax.legend(loc='best', frameon=True)

# Add a colorbar to show the gradient
cbar = fig.colorbar(sm, ax=ax)
cbar.set_label('Sample Order')

# Ensure the 'plots' directory exists
plots_dir = Path('plots_storms')
plots_dir.mkdir(parents=True, exist_ok=True)

# Save the figure
plt.savefig(plots_dir / 'carbon_endmembers_storm2.png')
plt.show()
```
