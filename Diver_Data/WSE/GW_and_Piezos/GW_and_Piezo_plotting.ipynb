```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
from matplotlib.ticker import StrMethodFormatter

directory = 'C:/Users/huck4481/Documents/GitHub/La_Jara/Diver_Data/WSE/GW_and_Piezos/data'
WSE_dict = {}

for filename in os.listdir(directory): # iterate over all files in GW_and_Piezos folder
    if filename.endswith('.csv'): # checking if its a csv
        # construct the full file path
        file_path = os.path.join(directory, filename)
        df = pd.read_csv(file_path)
        df['Date_Time'] = pd.to_datetime(df['Date_Time'])
        # remove the '.csv' from the filename to use as the dictionary key
        key = filename[:-4]
        # store the data frame in the dictionary
        WSE_dict[key] = df
```

### 2021 Plots: 

Downstream

```python
# Set Seaborn style
sns.set_theme(style="whitegrid", palette="pastel")

# Create subplots with a white background
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 9), sharex=True, facecolor='white')

# Plot for Location at P1
sns.lineplot(data=WSE_dict['GW_down_left_2021'], x='Date_Time', y='P1', color='#555555', linestyle='-', ax=axes[0])
sns.lineplot(data=WSE_dict['GW_down_right_2021'], x='Date_Time', y='P1', color='gray', linestyle=':', ax=axes[0])
sns.lineplot(data=WSE_dict['P1C_21_WSE'], x='Date_Time', y='WSE', color='blue', ax=axes[0])
axes[0].set_title('Water Elevations at Piezometer 1')
axes[0].set_ylabel('Elevation (m)')
axes[0].yaxis.set_major_formatter(StrMethodFormatter('{x:,.1f}')) 

# Plot for Location at P2
sns.lineplot(data=WSE_dict['GW_down_left_2021'], x='Date_Time', y='P2', color='#555555', linestyle='-', ax=axes[1])
sns.lineplot(data=WSE_dict['GW_down_right_2021'], x='Date_Time', y='P2', color='gray', linestyle=':', ax=axes[1])
sns.lineplot(data=WSE_dict['P2C_21_WSE'], x='Date_Time', y='WSE', color='blue', ax=axes[1])
axes[1].set_title('Water Elevations at Piezometer 2')
axes[1].set_ylabel('Elevation (m)')
axes[1].yaxis.set_major_formatter(StrMethodFormatter('{x:,.1f}')) 

# Plot for Location at P3
sns.lineplot(data=WSE_dict['GW_down_left_2021'], x='Date_Time', y='P3', label='Left GWE', color='#555555', linestyle='-', ax=axes[2], legend=False)
sns.lineplot(data=WSE_dict['GW_down_right_2021'], x='Date_Time', y='P3', label='Right GWE', color='gray', linestyle=':', ax=axes[2], legend=False)
sns.lineplot(data=WSE_dict['P3C_21_WSE'], x='Date_Time', y='WSE', label='Water Surface', color='blue', ax=axes[2], legend=False)
axes[2].set_title('Water Elevations at Piezometer 3')
axes[2].set_ylabel('Elevation (m)')
axes[2].set_xlabel('Time')
axes[2].yaxis.set_major_formatter(StrMethodFormatter('{x:,.1f}')) 

# Adjust layout
plt.tight_layout()
fig.suptitle('Piezometer and Ground Water Elevations - Downstream 2021', fontsize=16, y=1.02)
fig.legend(loc='lower center', bbox_to_anchor=(0.5, -0.04), fancybox=True, shadow=True, ncol=3)
plt.savefig('GW_piezo_down_2021.png', dpi=300, bbox_inches='tight')
plt.show()
```
