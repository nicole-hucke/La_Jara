```python
import numpy as np
import matplotlib.pyplot as plt

# create an array of values from 0 to 2*pi (360 degrees)
t = np.linspace(0, 2 * np.pi, 1000)

# define equation for a heart shape
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# plot the heart
plt.figure(facecolor='white')
plt.plot(x, y, color='red', linewidth='10')
plt.axis('equal') # Para que se vea simétrico 
plt.title("Happy Valentine's Day!")
plt.text(-12, 0, 'Code to Communicate', fontsize = 15, color='red')
plt.savefig('sanvalentin.png', dpi=300)
plt.show()

```
