## Correlation between Jemez and La Jara discharges during snowmelt season

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from scipy.signal import correlate, find_peaks

# Load data from CSV files
jemez = pd.read_csv('JemezRiverData.csv')
lajara = pd.read_csv('LaJara.csv')

# Convert date column to datetime object
jemez['date'] = pd.to_datetime(jemez['date'])
jemez = jemez.drop('Unnamed: 2', axis=1)
lajara['date'] = pd.to_datetime(lajara['date'])

# Set date as index
jemez.set_index('date', inplace=True)
lajara.set_index('date', inplace=True)
```

```python
# Calculate the daily average of the 'cfs' column in La Jara
lajara = lajara.resample('D').mean()
lajara = lajara.reset_index() # reset the index and rename columns
lajara.set_index('date', inplace=True) # set the date column as the index

# Clean La Jara
lajara = lajara.replace(-9999, np.nan)
lajara = lajara.mask(lajara < 0)
lajara = lajara.dropna()

# Drop the older years of jemez river
jemez = jemez[jemez.index.year >= 2017]
```

```python
# Filter data for spring months (March, April, May)
spring_months = [3, 4, 5]
jemez_spring = jemez[jemez.index.month.isin(spring_months)]
lajara_spring = lajara[lajara.index.month.isin(spring_months)]
```

```python
# Group data by year
jemez_grouped = jemez.groupby(jemez.index.year)

# Plot each year's discharge time series
plt.figure(figsize=(10, 6))
for year, data in jemez_grouped:
    plt.plot(data.index, data['cfs'], label=str(year))

# Add plot labels and legend
plt.xlabel('Date')
plt.ylabel('Discharge (cfs)')
plt.title('Jemez River Discharge Time Series')
plt.legend()
plt.show()
```

### Discharges in 2017

```python
# Filter jemez_spring for year 2017
jemez_spring_2017 = jemez_spring[jemez_spring.index.year == 2017]
lajara_spring_2017 = lajara_spring[lajara_spring.index.year == 2017]

# Create a figure with 6 subplots
fig, ax1 = plt.subplots(1, 1, figsize=(10, 5), sharex=True)
ax1.plot(lajara_spring_2017.index, lajara_spring_2017['cfs'], label='La Jara', color='blue')
ax1.set_ylabel('Discharge La Jara (cfs)')
ax2 = ax1.twinx()
ax2.plot(jemez_spring_2017.index, jemez_spring_2017['cfs'], label='Jemez', color='red')
ax2.set_ylabel('Discharge Jemez (cfs)')


# Set the x-axis locator and formatter
ax1.xaxis.set_major_locator(mdates.MonthLocator())
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%b'))

# Show the plot
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')
ax1.set_xlabel('Month')
ax1.set_title('Spring Discharge - 2017')
plt.show()
```
