```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load the FLAC audio file
file_path = "test_audio2.flac"
audio, sr = librosa.load(file_path, sr=None, mono=False)

# Separate the left and right channels
left_channel = audio[0]
right_channel = audio[1]

fig, axs = plt.subplots(2, 1, figsize=(15, 8))  # 2 rows, 1 column, figure size 10x8

# Plot the left channel in the first subplot
librosa.display.waveshow(left_channel, sr=sr, ax=axs[0])
axs[0].set_title('Left Gain Soundwave')
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Amplitude')

# Plot the right channel in the second subplot
librosa.display.waveshow(right_channel, sr=sr, ax=axs[1])
axs[1].set_title('Right Gain Soundwave')
axs[1].set_xlabel('Time (s)')
axs[1].set_ylabel('Amplitude')

plt.tight_layout()  # Ensure subplots don't overlap
plt.savefig('ejemplo_angel.png', dpi=300)
plt.show()
```
