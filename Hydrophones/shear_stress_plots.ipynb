## Shear Stress Hydrograph for each storm

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import shear stress data, making first column date_time and the index
tau_2021 = pd.read_csv('mobile_shear_stress_2021.csv', index_col=0, parse_dates=True)
tau_2022 = pd.read_csv('mobile_shear_stress_2022.csv', index_col=0, parse_dates=True)
tau_2023 = pd.read_csv('mobile_shear_stress_2023.csv', index_col=0, parse_dates=True)
# give the second column a name
tau_2021.columns = ['tau']
tau_2022.columns = ['tau']
tau_2023.columns = ['tau']
```

### Storm 1

```python
# Set your desired date (as a string in the format 'MM/DD/YYYY')
desired_date = '07/23/2021'

# Filter tau_2021 to include only the data for the desired date
tau_filtered = tau_2021[tau_2021.index.date == pd.to_datetime(desired_date).date()]

# Plot shear stress for the desired date
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
plt.title(f'Shear Stress on {desired_date}', fontsize=16)
plt.plot(tau_filtered.index, tau_filtered['tau'], label=f'{desired_date}')
plt.xlabel('Time', fontsize=14)
plt.ylabel('Shear Stress (Pa)', fontsize=14)
plt.legend()
plt.show()
```
