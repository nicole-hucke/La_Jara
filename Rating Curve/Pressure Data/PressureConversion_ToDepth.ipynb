```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from a CSV file
data = pd.read_csv('P3C-SM21.csv')

# Get the parameters
pressure = data['Pressure']   # Pressure reading in cmH2O
temp = data['Temperature']    # Water temperature in degrees Celsius
barom = data['Barom']  # Atmospheric pressure in cmH2O
d_transducer = 0.38   # Depth of the pressure transducer in m

# Convert the pressure from cmH2O to Pascals
barom_pa = barom * 98.0665
pressure_pa = pressure * 98.0665

# Calculate the water density at the measured temperature
rho = 1000 / (1 + 0.00021 * (temp - 4))

# Calculate the water depth
h = (pressure_pa - barom_pa) / (rho * 9.81) + (0.5 - d_transducer)
```

```python
fig = plt.figure(figsize=(30,7))
ax = fig.add_subplot(1,1,1)

ax.plot(data['Date/time'], h)
ax.set_xticks([]) 
ax.set_xlabel('Time')
ax.set_ylabel('Water Depth (m)')
plt.show()
```

```python
df = pd.DataFrame()
df['Date'] = data['Date/time'];
df['Depth (m)'] = h

df.to_csv('depthP3C_SM21.csv', index=False)
```
