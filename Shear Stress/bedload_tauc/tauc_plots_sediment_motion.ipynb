```python
# import stuff
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# read in the data, parse the dates, and set the index
tau_m = pd.read_csv('tau_mobile_preliminary.csv', parse_dates=['Date_Time'], index_col='Date_Time')
# separate by season
spring = tau_m['2023-03-01':'2023-06-21']
summer = tau_m['2023-06-22':'2023-09-21']
# make index a datetime objectc
spring.index = pd.to_datetime(spring.index)
summer.index = pd.to_datetime(summer.index)
```

#### Calculate 	$\tau$ / $\tau_c$
Tauc was calculated in the $\tau_c$ excel sheet

```python
# create copies of the slices 
spring = tau_m[(tau_m.index >= '2023-03-01') & (tau_m.index <= '2023-06-21')].copy()
summer = tau_m[(tau_m.index >= '2023-06-21') & (tau_m.index <= '2023-09-21')].copy()

# spring tau/tauc* by sizes
spring.loc[:, '2'] = spring['tau'] / 29.426
spring.loc[:, '1.4'] = spring['tau'] / 37.516
spring.loc[:, '1'] = spring['tau'] / 47.177
spring.loc[:, '0.7'] = spring['tau'] / 60.148
spring.loc[:, '0.5'] = spring['tau'] / 75.637
spring.loc[:, '0.2'] = spring['tau'] / 141.166

# summer u*/ws values (settling velocities)
summer.loc[:, '2'] = summer['tau'] / 29.426
summer.loc[:, '1.4'] = summer['tau'] / 37.516
summer.loc[:, '1'] = summer['tau'] / 47.177
summer.loc[:, '0.7'] = summer['tau'] / 60.148
summer.loc[:, '0.5'] = summer['tau'] / 75.637
summer.loc[:, '0.2'] = summer['tau'] / 141.166
```

Spring Plot

```python
# plot for sping u*/ws values (settling velocities) for all sizes
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
plt.plot(spring['2'], color='darkslategray', label='2 mm')
plt.plot(spring['1.4'], color='darkcyan', label='1.4 mm')
plt.plot(spring['1'], color='cadetblue', label='1 mm')
plt.plot(spring['0.7'], color='skyblue', label='0.7 mm')
plt.plot(spring['0.5'], color='lightblue', label='0.5 mm')
plt.plot(spring['0.2'], color='powderblue', label='0.2 mm')

plt.axhline(y=1, color='k', linestyle='--')

plt.ylabel(r'$\tau$ / $\tau_c$')
plt.xlabel('')
plt.legend()
plt.suptitle(r'Spring $\tau$ / $\tau_c$ by Size')
plt.tight_layout()
plt.show()
```

Summer Plot

```python
# plot for sping u*/ws values (settling velocities) for all sizes
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
plt.plot(summer['2'], color='darkslategray', label='2 mm')
plt.plot(summer['1.4'], color='darkcyan', label='1.4 mm')
plt.plot(summer['1'], color='cadetblue', label='1 mm')
plt.plot(summer['0.7'], color='skyblue', label='0.7 mm')
plt.plot(summer['0.5'], color='lightblue', label='0.5 mm')
plt.plot(summer['0.2'], color='powderblue', label='0.2 mm')

plt.axhline(y=1, color='k', linestyle='--')

plt.ylabel(r'$\tau$ / $\tau_c$')
plt.xlabel('')
plt.legend()
plt.suptitle(r'Summer $\tau$ / $\tau_c$ by Size')
plt.tight_layout()
plt.show()
```
